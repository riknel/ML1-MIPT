```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.cross_validation import cross_val_score, train_test_split
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from tqdm import tqdm
import sklearn
from imblearn.over_sampling import RandomOverSampler
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train.csv', encoding='latin')
df.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

Nan-ов нигде нет, хорошо

Самый хороший результат на кросс-валидации дал SVM c ядром rbf, поэтому будем использовать его

```python
features_train = np.array(df.iloc[:,2:])
y_train = np.array(df.label)
features_test = np.array(df_test.iloc[:,1:])
```

### SVM 

```python
clf = svm.SVC(kernel='rbf', random_state=0, gamma=0.001, probability=True)
clf.fit(features_train, y_train)
predictions_proba = clf.predict_proba(features_test)
```

```python
predictions_proba
```

```python
predictions = predictions_proba[:,1]
predictions[predictions>0.5] = 1
predictions[predictions<=0.5] = 0
```

Записываем результат

```python
answers = pd.read_csv('sampleSubmission.csv')
answers['label'] = list(map(int, np.array(predictions)))
answers.to_csv('attempt1.csv',  index=False)
answers.head()
```
