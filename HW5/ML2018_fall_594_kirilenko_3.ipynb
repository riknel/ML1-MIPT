## –ó–∞–¥–∞—á–∞ 1
–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ç–æ—Ä –≤ –∑–∞–¥–∞—á–µ –ª–∏–Ω–µ–π–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–º–µ–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω—ã–π —Å–º—ã—Å–ª –∞–ø—Ä–∏–æ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–µ–π. –ö–∞–∫–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–∞–¥–∞—é—Ç ùëô1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ç–æ—Ä –∏ ùëô2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ç–æ—Ä?


–†–∞—Å–ø–∏—à–µ–º –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏–µ —Å –∞–ø—Ä–∏–æ—Ä–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é $p(w)$: <br>
$$L(X) = \prod_{i=1}^n P(y_i|x_i) p(w) = \prod_{i:y_i=1} P(y_i|x_i) \prod_{i:y_i=0} (1 - P(y_i|x_i)) * p(w)$$ <br>
$$Ln L(X) = \sum_{i=1}^n y_i \ln(P(1|x_i)) + (1 - y_i) \ln(1 - P(1|x_i)) + \ln(p(w)) = \sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) + \ln p(w) \rightarrow max$$ <br>
–¢–æ –µ—Å—Ç—å –ø–æ–ª—É—á–∏–º : <br>
$$-\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) - \ln p(w) \rightarrow min$$ <br>
–¢–æ –µ—Å—Ç—å –∫–∞–∫ —Ä–∞–∑ –ø–æ–ª—É—á–∏–ª–∏ –Ω–µ–∫—É—é —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é: –∫ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—É –æ—à–∏–±–∫–∏ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ—Ç –≤–µ—Å–æ–≤. <br>

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ –∞–ø—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $N(0, \sigma^2I)$ –∏ –ø–æ–ª—É—á–∏–º: <br>
$$-\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) - \ln (\frac{1}{(2\pi)^{\frac{n}{2}} (\sigma^2)^\frac{1}{2}} e^{-\frac{1}{2}w^T \frac{1}{\sigma^2}w} )= -\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) + \frac{1}{2} \ln\sigma^2(2\pi)^n + \frac{1}{2\sigma^2}w^T w \rightarrow min$$
–¢–æ –µ—Å—Ç—å –ø–æ–ª—É—á–∏–º: <br>
$$-\sum_{i=1}^n y_i \ln(P(1|x_i)) + (1 - y_i) \ln((1 - P(1|x_i))) + \frac{1}{2\sigma^2}||w||\rightarrow min$$
–≠—Ç–æ L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è <br>
–¢–µ–ø–µ—Ä—å —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –õ–∞–ø–ª–∞—Å–∞ —Å $\beta = 0$ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞–ø—Ä–∏–æ—Ä–Ω–æ–≥–æ: <br>
$$-\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) - \ln \prod_{i=1}^n \frac{\alpha}{2} e^{-\alpha|w_i|} =  -\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) - n \ln \frac{\alpha}{2} + \alpha \sum_{i=1}^n |w_i|\rightarrow min$$
–¢–æ –µ—Å—Ç—å –ø–æ–ª—É—á–∏–º: <br>
$$-\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) + \alpha \sum_{i=1}^n |w_i| = -\sum_{i=1}^n y_i \ln(P(1|x_i) ) + (1 - y_i) \ln((1 - P(1|x_i))) + \alpha ||w||_1\rightarrow min$$

## –ó–∞–¥–∞—á–∞ 2

–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ –≤ —Å–ª—É—á–∞–µ –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–æ–π –≤—ã–±–æ—Ä–∫–∏ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª, –º–∞–∫—Å–∏–º–∏–∑–∏—Ä—É–µ–º—ã–π –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –¥–æ—Å—Ç–∏–≥–∞–µ—Ç —Å–≤–æ–µ–≥–æ –º–∞–∫—Å–∏–º—É–º–∞ –≤ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ (–ø—Ä–∏ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –≤–µ–∫—Ç–æ—Ä–∞ –≤–µ—Å–æ–≤), –∞ –∑–Ω–∞—á–∏—Ç –¥–ª—è –ø—Ä–æ—Ü–µ—Å—Å–∞ –æ–±—É—á–µ–Ω–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–æ–±–∞–≤–∏—Ç—å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–æ–Ω–Ω—ã–π —á–ª–µ–Ω.


–ü—É—Å—Ç—å —É –Ω–∞—Å –µ—Å—Ç—å –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–∞—è –≤—ã–±–æ—Ä–∫–∞ –∏ –ø—É—Å—Ç—å –≤–µ–∫—Ç–æ—Ä $\overline w$ –∏–¥–µ–∞–ª—å–Ω–æ –æ—Ç–¥–µ–ª—è–µ—Ç –≤—ã–±–æ—Ä–∫—É (–±–µ–∑ –æ—à–∏–±–æ–∫). –¢–æ–≥–¥–∞ $\forall \alpha \;\alpha \overline w$ —Ç–æ–∂–µ –∏–¥–µ–∞–ª—å–Ω–æ —Ä–∞–∑–¥–µ–ª—è–µ—Ç –≤—ã–±–æ—Ä–∫—É. –¢–æ–≥–¥–∞
$$Q(\alpha \overline w) = -\sum_{i=1}^l y_i \ln p_i + (1 - y_i) \ln (1 - p_i) = -\sum_{i=1}^l y_i \ln \frac{1}{1 + e^{-<\alpha \overline w,x_i>}} + (1 - y_i) \ln (\frac{e^{-<\alpha \overline w,x_i>}}{1 + e^{-<\alpha \overline w,x_i>}}) = -\sum_{i=i}^l ln(1 + e^{-\alpha y_i <\overline w, x_i>}) \rightarrow min$$
–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ –º—ã —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–ª–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–µ $\overline w$, –ø–æ—ç—Ç–æ–º—É $y_i <\overline w, x_i> $ —è–≤–ª—è–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–π –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ–π, –∑–∞–≤–∏—Å—è—â–µ–π –æ—Ç –Ω–æ–º–µ—Ä–∞ i. <br> –¢–æ–≥–¥–∞ 
$$Q =  -\sum_{i=i}^l ln(1 + e^{-\alpha c_i}) \rightarrow min$$
–û—Ç—Å—é–¥–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–∏ $\alpha \rightarrow \infty \;$ $e^{-\alpha c_i}$ –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ 0 –∏ –ø–æ—ç—Ç–æ–º—É $Q \rightarrow 0$. <br>
–í –∏—Ç–æ–≥–µ $|\alpha \overline w| \rightarrow \infty$ <br>
–ü—Ä–∏ —ç—Ç–æ–º, –∑–∞–º–µ—Ç–∏–º, —á—Ç–æ –µ—Å–ª–∏ –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é, —Ç–æ —Ç–∞–∫–∏—Ö –ø—Ä–æ–±–ª–µ–º –Ω–µ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç, —Ç–∞–∫ –∫–∞–∫ –º—ã –∫—Ä–æ–º–µ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª–∞ Q –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º –≤–µ—Å–∞ w –∏ –Ω–µ –¥–∞–µ–º –∏–º —É–π—Ç–∏ –≤ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç—å .

## –ó–∞–¥–∞—á–∞ 3

–ó–∞–¥–∞–Ω–∞ –≤—ã–±–æ—Ä–∫–∞ –æ–±—ä–µ–∫—Ç–æ–≤, –∫–∞–∂–¥—ã–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –æ–¥–Ω–æ–º—É –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Å–æ–≤ $Y = \{1; -1\}$. –í –Ω–µ–∫–æ—Ç–æ—Ä–æ–º —Å–ø—Ä—è–º–ª—è—é—â–µ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∑–∞–¥–∞–≤–∞–µ–º–æ–º —è–¥—Ä–æ–º $K_1(x_1; x_2)$, —ç—Ç–∞ –≤—ã–±–æ—Ä–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–æ–π, –∞ –≤ –Ω–µ–∫–æ—Ç–æ—Ä–æ–º —Å–ø—Ä—è–º–ª—è—é—â–µ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∑–∞–¥–∞–≤–∞–µ–º–æ–º —è–¥—Ä–æ–º $K_2(x_1; x_2)$, –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–æ–π. –°—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ —Å–ø—Ä—è–º–ª—è–µ–º–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, –∑–∞–¥–∞–≤–∞–µ–º–æ–µ —è–¥—Ä–æ–º $K(x_1; x_2) = K_1(x_1; x_2) + K_2(x_1; x_2)$, –≤ –∫–æ—Ç–æ—Ä–æ–º —ç—Ç–∞ –≤—ã–±–æ—Ä–∫–∞ –±—É–¥–µ—Ç –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–æ–π?


–ü—É—Å—Ç—å –ø–µ—Ä–≤–æ–µ —è–¥—Ä–æ –∑–∞–¥–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π $\phi$, —Ç–æ –µ—Å—Ç—å $K_1(w, x) = <\phi(w), \phi(x)>$ –∏ –ø—É—Å—Ç—å $w, w_0$ —Ä–∞–∑–¥–µ–ª—è—é—Ç –≤—ã–±–æ—Ä–∫—É –≤ —ç—Ç–æ–º —Å–ø—Ä—è–º–ª—è—é—â–µ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, —Ç–æ –µ—Å—Ç—å $K_1(w, x) - w_0 = <\phi(w), \phi(x)> - w_0 = 0$. <br>
–í—Ç–æ—Ä–æ–µ —è–¥—Ä–æ –∑–∞–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏–µ–π $\psi$ : $K_2(w, x) = <\psi(w), \psi(x)>$. <br>
–¢–µ–ø–µ—Ä—å –∑–∞–º–µ—Ç–∏–º, —á—Ç–æ $ <\phi(w), \phi(x)> = \phi(w)^T \phi(x) = \sum \phi(w_i) \phi(x_i) \;\;$ –∏ $ \;\;<\psi(w), \psi(x)> = \psi(w)^T \psi(x) = \sum \psi(w_i) \psi(x_i)$ <br>
–¢–æ–≥–¥–∞ $K_1(w, x) + K_2(w, x) = \sum \phi(w_i) \phi(x_i) + \sum \psi(w_i) \psi(x_i) = \phi(w)^T \phi(x) + \psi(w)^T \psi(x) = (\matrix{\phi(w)\\ \psi(w)})^T \; (\matrix{\phi(x)\\ \psi(x)})$ <br>
–¢–æ –µ—Å—Ç—å –ø–æ–ª—É—á–∏–º, —á—Ç–æ –µ—Å–ª–∏ $w, w_0$ —Ä–∞–∑–¥–µ–ª—è–µ—Ç –≤—ã–±–æ—Ä–∫—É –≤ —Å–ø—Ä—è–º–ª—è—é—â–µ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∑–∞–¥–∞–≤–∞–µ–º–æ–º $K_1(w, x)$, —Ç–æ ($\matrix {w \\ 0_{n}}), \;w_0$ —Ä–∞–∑–¥–µ–ª—è–µ—Ç –≤—ã–±–æ—Ä–∫—É –≤ —Å–ø—Ä—è–º–ª—è—é—â–µ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∑–∞–¥–∞–≤–∞–µ–º–æ–º $K_1(w, x) + K_2(w, x)$, —Ç–∞–∫ –∫–∞–∫: $K_1(w, x) + K_2(0_n, x) - w_0 = (\matrix{\phi(w)\\ \psi(0_n)})^T \; (\matrix{\phi(x)\\ \psi(x)}) - w_0 = <\phi(w), \phi(x)> - w_0 = 0$ <br>
–¢–æ –µ—Å—Ç—å –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, –∫–æ—Ç–æ—Ä–æ–µ –∑–∞–¥–∞–µ—Ç—Å—è $K_1(w, x) + K_2(w, x)$, –≤—ã–±–æ—Ä–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º–æ–π.
