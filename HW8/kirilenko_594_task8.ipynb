**Вопрос 1**: Чем LSTM лучше/хуже чем обычная RNN?

В RNN затухают градиенты, то есть чем дальше старое слово от текущего слова, тем меньше оно влияет на новое слово. Поэтому с большими текстами у RNN возникают проблемы. LSTM решает эту проблему (на каждом шаге "фильтруем" и оставляем то, что нужно). Однако LSTM все равно не сможет увидеть слово, которой слишком далеко. Недостатком LSTM является то, что он значительно долше обучается в силу его сложной стректуры.

**Вопрос 2**:  Выпишите производную $\frac{d c_{n+1}}{d c_{k}}$ для LSTM http://colah.github.io/posts/2015-08-Understanding-LSTMs/, объясните формулу, когда производная затухает, когда взрывается?

$$\frac{d c_{n+1}}{d c_{k}} = \prod_{i=k+1}^{n+1}\frac{d c_{i+1}}{d c_{i}} = \prod_{i=k+1}^{n+1}diag(f_i) = diag\left(\prod_{i=k+1}^{n+1}f_i\right)$$

$diag(f_i)$ - матрица, на диагонали которой стоят координаты вектора $f_i$, а остальные элементы 0 <br>
произведение $f_i$ - произведение векторов поэлементно. 

Тогда получим, что при $f_i^j > 1$ производная по $j$ координате взрывается, а при $f_i^j < 1$ затухает.

**Вопрос 3**: Зачем нужен TBPTT почему BPTT плох?

В TBPTT мы вместо того, чтобы считать градиенты по всему ряду, считаем только градиенты по параметрам, которые находятся на определенном расстоянии. Во-первых, это ускоряет обучение. Во-вторых, решает проблему затухания градиента. И иногда помогает избежать переобучения. В BPTT же мы считаем градиенты по всем параметрам и из-за этого, мы можем пытаться аптейтить параметры на совсем небольшие изменения и кроме того, появляется проблема затухания градиента.

1) Machine Translation/ Summarization

CNN может использоваться на уровне слов для получения ембединга предложения. Изначально слова в тексте заебедины, например с помощью word2vec. Текст представляется в виде трехмерной матрицы (кол-во предложений в тексте, кол-во слов в предложении, размерность ембединга слова). Различными фильтрами размера (k,  размерность ембединга слова) и последующим применением пулинга получим представление предложения. После этого можно запустить обычный encoder-decoder. Encoder-у на вход подаются предложения и последним hidden state можно считать представлением документа. После работает decoder, которому в случае суммаризации подаются на вход предложения (возможно домноженные на вероятность войти в саммари), а в случае перевода истинные слова.

К исходной картинке применяется CNN и получается feature map, которая складывается внутри каждого канала и подается в RNN. RNN на каждом шаге возвращает слово для описания картинки и некоторую маску весов, где пиксели имеют больший вес, если RNN хочет обратить на них внимание на следующей итериции. Эта маска домножается на исходную feature map и подается на следующую итерацию RNN. Таким образом, RNN на каждом шаге обращает больше внимания на определенные пиксели и генерирует описание картинки.

**Вопрос 6**: Attention - что это такое, где применяют и как? Приведите пример использования на какой-нибудь задаче

Attention - механизм, выделяющий определенные части входных данных, например фрагменты текста или изображений,  для более детальной обработки на следующей итерации. Attention очень часто применяют в seq2seq моделях и в качестве объекта внимания обычно выступает выход последнего слоя кодирующей части. 
<br> 
Для простоты рассмотрим как он работает в модели encoder-decoder. Обозначим $h_i$ - скрытые состояния последнего слоя енкодера. attention для каждого скрытого состояния декодера $s_j$ суммирует скрытые состояния енкодера $h_i$ с какими-то коэффициентами для получения контекст вектора. Коэффициенты перед каждым $h_i$ определяютя как $f(h_i, s_j)$, где в качестве f обычно берут $dot(h_i, s_j)$ или $h_i^T W s_j$. Полученные коэффициенты для каждого $h_i$ после нормируются (берем softmax) и после суммирования $h_i$ с этими состояниями получается вектор контекста. 
Attention применяют в таким задачах как машинный перевод, суммаризация текста. В этих случаях attention работает описанным выше способом. <br>
Также attention можно применять и для картинок, например для image captioning. В данном случае механизм внимания будет указывать на определенные пиксели картинок, на которые модель хочет обратить внимание для генерации следующего слова.

### Data

Now we need to generate data. It will be dates in different text formats and in fixed output format.

```python
from faker import Faker
import random
from tqdm import tqdm
from babel.dates import format_date
import numpy as np
```

```python
fake = Faker()

FORMATS = ['short',
           'medium',
           'long',
           'full',
           'd MMM YYY', 
           'd MMMM YYY',
           'dd MMM YYY',
           'd MMM, YYY',
           'd MMMM, YYY',
           'dd, MMM YYY',
           'd MM YY',
           'd MMMM YYY',
           'MMMM d YYY',
           'MMMM d, YYY',
           'dd.MM.YY']

# change this if you want it to work with another language
LOCALES = ['en_US']
```

```python
def create_date():
    """
        Creates some fake dates 
        :returns: tuple containing human readable string, machine readable string, and date object
    """
    dt = fake.date_object()

    try:
        human_readable = format_date(dt, format=random.choice(FORMATS), locale=random.choice(LOCALES))

        case_change = random.choice([0,1,2])
        if case_change == 1:
            human_readable = human_readable.upper()
        elif case_change == 2:
            human_readable = human_readable.lower()
        # if case_change == 0, do nothing

        machine_readable = dt.isoformat()
    except AttributeError as e:
        return None, None, None

    return human_readable, machine_readable, dt
```

```python
def create_dataset(n_examples):
    """
        Creates a dataset with n_examples and vocabularies
        :n_examples: the number of examples to generate
    """
    human_vocab = set()
    machine_vocab = set()
    dataset = []

    for i in tqdm(range(n_examples)):
        h, m, _ = create_date()
        if h is not None:
            dataset.append((h, m))
            human_vocab.update(tuple(h))
            machine_vocab.update(tuple(m))

    human = dict(zip(list(human_vocab) + ['<unk>', '<pad>'], 
                     list(range(len(human_vocab) + 2))))
    inv_machine = dict(enumerate(list(machine_vocab) + ['<unk>', '<pad>']))
    machine = {v:k for k,v in inv_machine.items()}
 
    return dataset, human, machine, inv_machine
```

```python
def string_to_int(string, lenght, vocab):
    if len(string) > lenght:
        string = string[:lenght]
        
    rep = list(map(lambda x: vocab.get(x, '<unk>'), string))
    
    if len(string) < lenght:
        rep += [vocab['<pad>']] * (lenght - len(string))
    
    return rep
```

```python
def int_to_string(ints, inv_vocab):
    return [inv_vocab[i] for i in ints]
```

### Part 1: Simple NMT

Чтобы не расписывать и так простую модель, я ее нарисовала

![page 1](1.png)

```python
# :good-enouht:
ENCODER_UNITS = 32 # change me if u want
DECODER_UNITS = 32 # change me if u want
TIME_STEPS = 20 # change me if u want
M_VOCAB = len(machine_vocab)
```

```python
# input - [bs; in_time_len]
# output - [bs; out_time_len]; out_time_len=10

def model_simple_nmt(in_chars, out_chars):
    # RNN encoder -> hidden representation -> RNN decoder
    inputs = Input(shape=(TIME_STEPS,))
    reshape_input = Reshape(target_shape=(TIME_STEPS, 1))(inputs)
    
    # encoder
    lstm_f1, h_f, c_f = LSTM(units=ENCODER_UNITS, return_sequences=True, return_state=True)(reshape_input)
    lstm_b1, h_b, c_b = LSTM(units=ENCODER_UNITS, return_sequences=True, return_state=True, go_backwards=True)(reshape_input)
    concat_lstm1 = Concatenate(axis=2)([lstm_f1, lstm_b1])
    reshape_lstm1 = Reshape(target_shape=(TIME_STEPS, 2*ENCODER_UNITS))(concat_lstm1)
    concat_h = Concatenate(axis=1)([h_f, h_b])
    concat_c = Concatenate(axis=1)([c_f, c_b])
    repeat_h = Lambda(lambda x: K.repeat(x, TIME_STEPS))(concat_h)
    
    # decoder
    lstm_f2 = LSTM(units=2*ENCODER_UNITS, return_sequences=True)(repeat_h, initial_state=(concat_h, concat_c))
    lstm_b2 = LSTM(units=2*ENCODER_UNITS, return_sequences=True, go_backwards=True)(repeat_h, initial_state=(concat_h, concat_c))
    concat_lstm2 = Concatenate(axis=2)([lstm_f2, lstm_b2])
    reshape_lstm2 = Reshape(target_shape=(TIME_STEPS, 4*ENCODER_UNITS))(concat_lstm2)
    output = TimeDistributed(Dense(M_VOCAB, activation='softmax'))(reshape_lstm2)


    model = Model(input=[inputs], output=output)
    return model
```

### Part 2: All u need is attention

Here we use more complex idea that simple seq2seq: we're adding two explicit parts of our network - encoder and decoder (which is applied attention on). The explanatory picture for this idea is below:

The lower part of the network is encoding the input to some hidden intermediate representation and the upper part is decoing the hidвen represenataion into some readable output.

```python
# :good-enouht:
ENCODER_UNITS = 32 # change me if u want
DECODER_UNITS = 32 # change me if u want
TIME_STEPS = 20 # change me if u want
M_VOCAB = len(machine_vocab)
```

Вспомогательная функция для взятия определенного среза тензора

```python
def crop(dimension, start, end):
    # Crops a Tensor on a given dimension from start to end
    def func(x):
        if dimension == 0:
            return x[start: end]
        if dimension == 1:
            return x[:, start: end]
        if dimension == 2:
            return x[:, :, start: end]
        if dimension == 3:
            return x[:, :, :, start: end]
        if dimension == 4:
            return x[:, :, :, :, start: end]
    return Lambda(func)
```

В attention-е для нахождения $\alpha_{ti}$, на которые домножаются hidden state-ы енкодера, я выбрала функцию $h_iWs_t$. <br>
А так, в общем обычный encoder - decoder. Разница с предыдущим здесь в том, что теперь я подаю на вход декодеру ответы (истинные во время обучения и предсказанные во время теста). Для этого нужно будет чуть позже видоизменить predict, то есть теперь мы будем предсказывать по одному следующему слову и подавать это слова следующему блоку LSTM-а декодера.

После decoder-a проходимся for-ом по полученным в декодере hidden state-ам, находим $\alpha_{ti}$, суммируем hidden state-ы енкодера с этими весами, конкатенируем с текущим hidden state-ом декодера и получаем вероятности, применив dense.

```python
def model_attention_nmt(in_chars, out_chars):
    # RNN encoder -> hidden representation -> RNN decoder
    
    inputs = Input(shape=(TIME_STEPS,))
    targets = Input(shape=(TIME_STEPS,))
    reshape_input = Reshape(target_shape=(TIME_STEPS, 1))(inputs)

    # encoder
    lstm_f1, h_f, c_f = LSTM(units=ENCODER_UNITS, return_sequences=True, return_state=True)(reshape_input)
    lstm_b1, h_b, c_b = LSTM(units=ENCODER_UNITS, return_sequences=True,return_state=True, go_backwards=True)(reshape_input)
    concat_lstm1 = Concatenate(axis=2)([lstm_f1, lstm_b1])
    reshape_lstm1 = Reshape(target_shape=(TIME_STEPS, 2*ENCODER_UNITS))(concat_lstm1)
    concat_h = Concatenate(axis=1)([h_f, h_b])
    concat_c = Concatenate(axis=1)([c_f, c_b])
    repeat_h = Lambda(lambda x: K.repeat(x, TIME_STEPS))(concat_h)

    dense_h = TimeDistributed(Dense(2*ENCODER_UNITS, activation='softmax'))(reshape_lstm1)

    # define dense layer for applying on each step
    dense = Dense(M_VOCAB, activation = 'softmax')

    # add initial entrance
    zeros = Lambda(lambda x: x[:, 0] * 0)(concat_h)
    zeros = Reshape((1, ))(zeros)
    concat = Concatenate(axis=1)([zeros, targets])
    new_targets = Reshape(target_shape=(TIME_STEPS+1, 1))(concat)

    # decoder
    lstm_2 = LSTM(units=2*ENCODER_UNITS, return_sequences=True)(new_targets, initial_state=(concat_h, concat_c))
    reshape_lstm2 = Reshape(target_shape=(TIME_STEPS+1, 2*ENCODER_UNITS))(lstm_2)

    # attention for zero hidden state
    h_0 = crop(1, 0, 1)(reshape_lstm2)
    h_0 = Reshape(target_shape=(2*ENCODER_UNITS,))(h_0)
    content = Lambda(lambda x: K.mean(x, axis=1))(reshape_lstm1)
    concat_content = Concatenate(axis=1)([content, h_0])
    # result for zero step
    res_0 = dense(concat_content)
    res_0 = Reshape(target_shape=(1, M_VOCAB))(res_0)
    result = [res_0]

    # attention for other hidden states
    for i in range(1, TIME_STEPS):
        h_i = crop(1, i, i + 1)(reshape_lstm2)
        h_i = Reshape(target_shape=(2*ENCODER_UNITS,))(h_i)
        h_prev = crop(1, i - 1, i)(reshape_lstm2)
        mul_alpha = Multiply()([dense_h, h_prev]) # reshape_lstm1 -> dense_h
        sum_alpha = Lambda(lambda x: K.sum(x, axis=2), output_shape=lambda s: (s[0], s[1]))(mul_alpha)
        alpha = Activation('softmax')(sum_alpha)
        alpha_repeat = Lambda(lambda x: K.repeat(x, 2*ENCODER_UNITS))(alpha)
        alpha_reshape =  Reshape((TIME_STEPS, 2*ENCODER_UNITS))(alpha_repeat)
        states_with_coef = Multiply()([reshape_lstm1, alpha_reshape])
        content_i = Lambda(lambda x: K.sum(x, axis=1), output_shape=lambda s: (s[0], s[2]))(states_with_coef)
        concat_content_i = Concatenate(axis=1)([content_i, h_i])
        res_i = dense(concat_content_i)
        res_i = Reshape(target_shape=(1, M_VOCAB))(res_i)
        result.append(res_i)

    # concatenate obtained result
    output = Concatenate(axis=1)(result)

    model = Model(input=[inputs, targets], output=output)
    return model
```

В отиличие от модели в статье я использую 4 GRU в енкодере (5 если считать двунаправленную GRU как 2) и 4 GRU в декодере. Кроме того в статье немного другая AttentionFunction, я для простоты взяла $h_iWs_t$.

```python
TIME_STEPS = 32
SENT_LEN = TIME_STEPS
ENCODER_DIM = 128
DECODER_DIM = 128
ATTENTION_DIM = 20
VOC_SIZE = 88
```

```python
def model_complex_nmt_tatoeba():
    english_input = Input(shape=(SENT_LEN, ), name='english')
    russian_input = Input(shape=(SENT_LEN, ), name='russian')
    english = Reshape(target_shape=(SENT_LEN, 1))(english_input)
    russian = Reshape(target_shape=(SENT_LEN, 1))(russian_input)

    # Encoder
    # first bi-gru
    gru_f1, h_f1 = GRU(units=ENCODER_DIM, return_sequences=True, return_state=True)(english)
    gru_b1, h_b1 = GRU(units=ENCODER_DIM, return_sequences=True, return_state=True, go_backwards=True)(english)
    reshape_gru_f1 = Reshape(target_shape=(SENT_LEN, ENCODER_DIM))(gru_f1)
    reshape_gru_b1 = Reshape(target_shape=(SENT_LEN, ENCODER_DIM))(gru_b1)
    concat_gru_1 = Concatenate(axis=2)([reshape_gru_f1, reshape_gru_b1])
    concat_h_1 = Concatenate(axis=1)([h_f1, h_b1])

    # second gru
    gru_2, h_2 = GRU(units=2*ENCODER_DIM, return_sequences=True, return_state=True)(concat_gru_1)
    reshape_gru_2 = Reshape(target_shape=(SENT_LEN, 2*ENCODER_DIM))(gru_2)
    gru_3, h_3 = GRU(units=2*ENCODER_DIM, return_sequences=True, return_state=True)(reshape_gru_2)
    reshape_gru_3 = Reshape(target_shape=(SENT_LEN, 2*ENCODER_DIM))(gru_3)

    # residual connections
    addition = Add()([reshape_gru_2, reshape_gru_3])

    # forth gru
    gru_4, h_4 = GRU(units=2*ENCODER_DIM, return_sequences=True, return_state=True)(addition)
    reshape_gru_4 = Reshape(target_shape=(SENT_LEN, 2*ENCODER_DIM))(gru_4)

    # first gru of decoder. we will apply it to every entrance separately 
    first_gru_cell= GRU(2*ENCODER_DIM, return_sequences=True, return_state=True)

    # dense function for finding alphas in context vector
    dense_h = TimeDistributed(Dense(2 * ENCODER_DIM, activation='softmax', use_bias=False))(reshape_gru_4)

    context_vectors = []
    next_input = []

    # attention for zero cell of gru
    # for zero cell context vector is mean of encoder hidden states
    context = Lambda(lambda x: K.mean(x, axis=1))(reshape_gru_4)
    context = Reshape(target_shape=(1, 2 * ENCODER_DIM))(context)
    new_input = Concatenate(axis=-1)([crop(1, 0, 1)(russian), context])
    context_vectors.append(context)

    # zero cell of first gru of decoder
    cell_0 = first_gru_cell(new_input, initial_state=concat_h_1)
    hidden = cell_0[1]
    seq = Reshape(target_shape=(1, 2 * ENCODER_DIM))(cell_0[0])
    next_input.append(seq)

    for i in range(1, TIME_STEPS):
        # find alphas
        mul_alpha = Multiply()([dense_h, hidden])
        sum_alpha = Lambda(lambda x: K.sum(x, axis=2), output_shape=lambda s: (s[0], s[1]))(mul_alpha)
        alpha = Activation('softmax')(sum_alpha)
        alpha_repeat = Lambda(lambda x: K.repeat(x, 2 * ENCODER_DIM))(alpha)
        alpha_reshape =  Reshape((TIME_STEPS, 2 * ENCODER_DIM))(alpha_repeat)

        # find context vector
        encoder_with_coefs = Multiply()([reshape_gru_4, alpha_reshape])
        context = Lambda(lambda x: K.sum(x, axis=1), output_shape=lambda s: (s[0], s[2]))(encoder_with_coefs)
        context = Reshape(target_shape=(1, 2 * ENCODER_DIM))(context)
        context_vectors.append(context)

        # apply gru to next entrance with attention vector
        new_input = Concatenate(axis=-1)([crop(1, i, i + 1)(russian), context])

        next_cell = first_gru_cell(new_input, initial_state=hidden)
        hidden = next_cell[1]
        seq = Reshape(target_shape=(1, 2 * ENCODER_DIM))(next_cell[0])
        next_input.append(seq)

    # second gru of decoder. we will apply it to every entrance separately 
    second_gru_cell = GRU(2*ENCODER_DIM, return_sequences=True, return_state=True)
    next_input_2 = []
    hidden = h_2
    for i in range(TIME_STEPS):
        new_input = Concatenate(axis=-1)([next_input[i], context_vectors[i]])
        next_cell = second_gru_cell(new_input, initial_state=hidden)
        hidden = next_cell[1]
        seq = Reshape(target_shape=(1, 2 * ENCODER_DIM))(next_cell[0])
        next_input_2.append(seq)

    # third gru of decoder. we will apply it to every entrance separately 
    third_gru_cell = GRU(2*ENCODER_DIM, return_sequences=True, return_state=True)
    next_input_3 = []
    hidden = h_3
    for i in range(TIME_STEPS):
        new_input = Concatenate(axis=-1)([next_input_2[i], context_vectors[i]])
        next_cell = third_gru_cell(new_input, initial_state=hidden)
        hidden = next_cell[1]
        seq = Reshape(target_shape=(1, 2 * ENCODER_DIM))(next_cell[0])
        next_input_3.append(seq)

    # forth gru of decoder. we will apply it to every entrance separately 
    forth_gru_cell = GRU(2*ENCODER_DIM, return_sequences=True, return_state=True)
    next_input_4 = []
    hidden = h_4
    for i in range(TIME_STEPS):
        new_input = Concatenate(axis=-1)([next_input_3[i], context_vectors[i]])
        next_cell = third_gru_cell(new_input, initial_state=hidden)
        hidden = next_cell[1]
        seq = Reshape(target_shape=(1, 2 * ENCODER_DIM))(next_cell[0])
        next_input_4.append(seq)

    hidden_output = Concatenate(axis=1)(next_input_4)
    result = TimeDistributed(Dense(units=VOC_SIZE, activation='softmax'))(hidden_output)
    
    model = Model(input=[english_input, russian_input], output=result)
    return model
```
